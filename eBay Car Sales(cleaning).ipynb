```python
# Viewing all the column names
all_columns = autos.columns
print(all_columns)
```

```python
# Renaming columns to accomodate the python Snake Case format
column_corrections = {"yearOfRegistration" : "registration_year",
"monthOfRegistration" : "registration_month", "notRepairedDamage" : "unrepaired_damage",
"dateCreated" : "ad_created", "dateCrawled" : "date_crawled", "name" : "name", "seller" : "seller", "offerType" : "offer_type", 
"price" : "price", "abtest" : "abtest", "vehicleType" : "vehicle_type", "gearbox" : "gearbox", "powerPS" : "power_ps", "model" : "model",
                      "odometer" : "odometer", "fuelType" : "fuel_type", "brand" : "brand", "nrOfPictures" : "picture_number", 
                     "postalCode" : "postal_code", "lastSeen" : "last_seen"}

autos.columns = autos.columns.map(column_corrections)

print(autos.columns)
```

```python
# Observing the dataframe after the above adjustments have been made
(autos.head())
```

```python
# Let us now look at the descriptive statistics for every column including the categorical ones
column_statistics = autos.describe(include = "all")
print(column_statistics)


# It is apparent that the price and distance_covered columns are numeric values stored as texts
# They need to be converted to numeric data types
# Also the number of pictures column seems to have the value 0 in all rows, we can drop it
# The seller and offer type columns are similar/almost similar, we can drop them as well as they lack useful info for analysis
```

### **General Observations of The Columns**
#### -It is apparent that the price and distance_covered columns are numeric values stored as texts
#### -They need to be converted to numeric data types
#### -Also the number of pictures column seems to have the value 0 in all rows, we can drop it
#### -The seller and offer type columns are similar/almost similar, we can drop them as well as they lack useful info for analysis

```python
                                
# Removing non-numeric characters from numeric columns (price column)
autos["price"] = autos["price"].str.replace("$","").str.replace(",", "")

# Converting the values to numeric data type
autos["price"] = autos["price"].astype(int)

# Show the first five columns
autos["price"].head()



```

```python

# Removing non-numeric characters from numeric columns (odometer column)
autos["odometer"] = autos["odometer"].str.replace("km", "").str.replace(",", "")

# Converting the values to numeric data type
autos["odometer"] = autos["odometer"].astype(int)

# Renaming the odometer column (without assignment)
autos.rename({"odometer" : "odometer_km"}, inplace = True, axis = 1)

# Show the first 5 rows
autos["odometer_km"].head()
```

```python
autos = autos.drop(["picture_number", "seller", "offer_type"], axis = 1)
autos
```

```python
# Analyzing the price column in details 
unique_price = autos["price"].unique().shape
print(unique_price)

price_describe = autos["price"].describe()
print(price_describe)

price_count = autos["price"].value_counts().sort_index(ascending = False).head(20)
print(price_count)

# It seems that there are outliers in the price column.
# prices increases steadily to $350,000 before jumping to unrealistic values of up to $99 million.
# We will need to remove prices that are above $350,000
# The rows with 0 as price account for 2% of the data which has no profound impact on the dataset
# These rows with 0 as price will also be removed 


```

### **Observations Made On the Price Column**

- Most cars are low-priced (typically between $1K and $7K) as the 75% percentile shows.
- But a few incorrect or extremely high prices like **$100 million** make the average much higher than it should be.
- The data is highly skewed as a few outliers are distorting the mean.
- Prices increases steadily to **$350,000** before jumping to unrealistic values of up to **$99 million.**
- We will need to remove prices that are above $350,000.
- The rows with 0 as price account for 2% of the data which has no profound impact on the dataset.
- These rows with 0 as price will also be removed.


```python
# Remove the outliers in the price column including the rows with 0 
price_accepted = autos["price"].between(1, 351000) # returns a boolean series
autos = autos[price_accepted] # returns a dataframe that meets the above condition
autos["price"].describe()
```

```python
# Exploring the registration_year column
autos["registration_year"].describe()

# The registration year of a car should tell us how old the vehicle is. However,  we find some strange and unrealistic entries:  
# The earliest year listed is 1000 centuries before cars even existed.  
# The latest year is 9999, which is thousands of years ahead in the future.  
# This suggests there are in the data that need to be fixed for accurate analysis. 
```

### **Observations on the Registration Year Column**
#### **-The registration year of a car should tell us how old the vehicle is. However,  we find some strange and unrealistic entries:**  
#### **-The earliest year listed is 1000 centuries before cars even existed.**  
#### **-The latest year is 9999, which is thousands of years ahead in the future.**  
#### **-This suggests there are in the data that need to be fixed for accurate analysis.** 

```python
# Historical context: Cars were mass-produced starting in the early 1900s.
# Current year: The dataset’s latest entry is from 2016, so no car can be registered after this.

autos = autos[autos["registration_year"].between(1910, 2016)]
autos["registration_year"].value_counts(normalize = True).head(10)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True) * 100

# German brands hold 4 of the top 5 spots
# These make up ~50% of all listings
# Volkswagen leads with double the listings of runners-up
# We'll analyze only brands with >5% share (significant players)
# Smaller brands (<5%) excluded for focused insights

common_brands = brand_counts[brand_counts > 5].index # index removes the percentages and returns just the names

print(common_brands)
```

```python
# Exploring and calculating the Price by Brand

brand_mean_prices = {}

for brand in common_brands:
   specific_brand =  autos[autos["brand"] == brand]
   mean_price = specific_brand["price"].mean()
   brand_mean_prices[brand] = int(mean_price)
print(brand_mean_prices)

# Premium Tier: Audi, BMW, and Mercedes-Benz command higher prices.

# Budget Tier: Ford and Opel are more affordable.

# Volkswagen’s catchy Spot: Priced midway, likely boosting its popularity by balancing cost and quality
```

### **Conclusions/Observations from Price by Brand Aggregation**
#### -Premium Tier: Audi, BMW, and Mercedes-Benz command higher prices.

#### -Budget Tier: Ford and Opel are more affordable.

#### -Volkswagen’s catchy Spot: Priced midway, likely boosting its popularity by balancing cost and quality


```python
# Exploring Mileage and Calculating the average mileage by brand
brand_average_mileage = {} # dictionary to store the average mileage by brand
for brand in common_brands:
    specific_brand = autos[autos["brand"] == brand]
    average_mileage = specific_brand["odometer_km"].mean()
    brand_average_mileage[brand] = int(average_mileage)
print(brand_average_mileage)
```

```python
# Storing Aggregate Data(stored in a dictionary) in a Dataframe

# Use the Pandas Series Constructor to Construct a Series Object for both dictionaries
bmp_series = pd.Series(brand_mean_prices)
bam_series = pd.Series(brand_average_mileage)

print(bmp_series)
print(bam_series)
```

```python
# Use the Pandas Dataframe Constructor to construct a dataframe using the first series object
bmp_dataframe = pd.DataFrame(bmp_series, columns = ["mean_price"])
bma_dataframe = pd.DataFrame(bam_series, columns = ["average_mileage"])

# Adding a column to the original dataframe
bmp_dataframe["average_mileage"] = average_mileage

print(bmp_dataframe)
```


### **Key Observations and Conclusions**
#### -Mileage consistency: Top brands vary by ≤10% in mileage

#### -Price disparity: Significant price gaps exist between brands

#### -Positive/direct trend; Pricier cars slightly show higher mileage
